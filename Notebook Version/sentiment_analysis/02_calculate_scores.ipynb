```python
from research_iet import process_all_posts, save_to_file
from os import path
import json
print("Libraries Imported!")
```

```python
# processing posts
pre_lockdown_file = open(path.join("..", 'data', 'pre_lockdown.json'),
                 'r', newline='', encoding="utf8")
in_lockdown_file = open(path.join("..", 'data', 'in_lockdown.json'),
                 'r', newline='', encoding="utf8")
```

```python
all_pre_posts = json.load(pre_lockdown_file)
all_in_posts = json.load(in_lockdown_file)
```

```python
all_pre_sentiments = process_all_posts(all_pre_posts)
all_in_sentiments = process_all_posts(all_in_posts)
print(f"Total posts {len(all_pre_posts) + len(all_in_posts)}")
```

```python
save_to_file(path.join('score', f'score_pre_lockdown.json'), all_pre_sentiments)
save_to_file(path.join('score', f'score_in_lockdown.json'), all_in_sentiments)
print('Scores calculated successfully üôè')
```
